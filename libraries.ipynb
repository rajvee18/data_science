# NumPy (Numerical Python)

#### Array

```python
import numpy as np

# numpy arrays are homogenous (same datatype)

#creating a 1d array

my_list=[1,2,3,4,5]
my_array=np.array(my_list)
print(my_array)
print(my_list)


#another way to create a 1d array

array1d = np.array([1,2,3,4,5])
print(array1d)


# 2d array

my_2d_array = np.array([[1,2,3],[2,3,4]])

print(my_2d_array)
print([[1,2,3],[2,3,4]])


# 3d array

my_3d_array = np.array([[[1,2],[3,4],[5,6]]])
print(my_3d_array)
```

#### np.zeroes(shape) and np.ones(shape)

```python
import numpy as np

zero_array=np.zeros((2,3))
print(zero_array)
```

```python
ones_array=np.ones((2,3))
print(ones_array)
```

#### Arithmetic operations

```python
array1=np.array([1,2,3,4,5])

print(array1/2)
print(array1+2)
print(array1-5)
print(array1*3)
```

#### Indexing

```python
my_2d_array = np.array([[1,2,3],[2,3,4]])

print(my_2d_array)
print(my_2d_array[1,:])   # (:) means 'entire'


my_3d_array = np.array([[[1,2],[3,4]],[[99,100],[5,6]]])

print(my_3d_array)
print(my_3d_array[1,0,1]) # [layer,row,coloumn]
 
```

#### Pandas Data Structures: 1. Series

```python
data=[1,2,3,4]

my_series=pd.Series(data)
print(my_series)

print(my_series[2])
```

# Matplotlib

#### Line plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,36,78,45,89]

plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel("numbers")
plt.ylabel("function")


```

#### Bar plot

```python
plt.bar(x,y,color="green")
plt.title("Bar plot")
plt.xlabel("numbers")
plt.ylabel("frequency")
```

#### Histogram

```python
data=[10,12,5,8,9,46,68,90,100,35,25]

plt.hist(data,bins=5,color="purple",edgecolor="black")
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("histogram")
plt.show()
```
