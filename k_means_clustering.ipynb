```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/home/code_wizard/coding_files/data_science/datasets/Unsupervised Machine Learning with K-Means Clustering in Python/Mall_Customers.csv')

x = df.iloc[:,3:5].values
x
```

### Using Elbow method to find optimal number of clusters

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=1)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)

plt.title("The Elbow Curve")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
```

### Training the dataset on KMeans with clusters=5

```python
kmeans1 = KMeans(n_clusters=5,init='k-means++',random_state=1)

y_kmeans = kmeans1.fit_predict(x)


print(np.concatenate((x,y_kmeans.reshape(len(y_kmeans),1)),1))
```
