### pandas.nunique() and unique() function

```python
unique_counts = df.nunique()
print(unique_counts)  #count the unique values in each coloumn

survived_unique = df['survived'].unique() # print the unique values
print(survived_unique)

sex_unique = df['sex'].unique()
print(sex_unique)
```

### pandas.value_counts() function

```python
# value_counts() function
## count the occurence of each unique value

gender_value_counts= df['sex'].value_counts()  
print(gender_value_counts)

survived_value_counts= df['survived'].value_counts()
print(survived_value_counts)
```

### pandas.groupby() function

```python
# to separate the passenger based on gender and then count how many have survived

grouped_data= df.groupby(['sex','survived'])['survived'].count()
print(grouped_data)

class_survived_counts= df.groupby(['pclass','survived'])['survived'].count()
print(class_survived_counts)

embarked_class_survived_counts= df.groupby(['embark_town','pclass','survived'])['survived'].count()
print(embarked_class_survived_counts)
```
